```python
import subprocess

def install_requirements(requirements_file):
  """
  Installs packages from a requirements.txt file,
  skipping installations that fail.

  Args:
    requirements_file: Path to the requirements.txt file.
  """
  with open(requirements_file, 'r') as f:
    for line in f:
      line = line.strip()
      if line and not line.startswith('#'):  # Skip empty lines and comments
        try:
          subprocess.check_call(['pip', 'install', line])
          print(f"Successfully installed: {line}")
        except subprocess.CalledProcessError:
          print(f"Failed to install: {line}")

# This:
install_requirements('/content/silic/requirements.txt')
```
